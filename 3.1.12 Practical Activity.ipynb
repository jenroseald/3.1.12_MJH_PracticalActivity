```python
# Import the packages
import pandas as pd
```

```python
# Import and read files into DFs
movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')
```

```python
# Check import has worked correctly
```

```python
print(movies.columns)
print(movies.shape)
```

```python
print(ott.columns)
print(ott.shape)
```

```python
# Merge the DFs on the left (movies is left)
```

```python
mov_ott = pd.merge(movies, ott, how='left', on='ID')
```

### Answer the 5 activity questions

```python
# Q1. Determine film release date and content rating (Age column)
```

```python
# Use 'Title', 'Year' and 'Age'
```

```python
movies_q1 = movies.pivot(index='Title', columns='Age', values='Year')
```

```python
print(movies_q1.columns)
print(movies_q1.shape)
```

```python
movies_q1
```

```python
# Q2 Determine the title of movies, directors and genres by 'Age'
```

```python
movies_q2 = movies.pivot(index='Title', columns='Age', values=['Directors', 'Genres'])
```

```python
print(movies_q2.columns)
print(movies_q2.shape)
```

```python
movies_q2
```

```python
# Q3 Determine Title, release year, language by 'Age'
```

```python
movies_q3 = movies.pivot(index='Title', columns='Age', values=['Year', 'Language'])
```

```python
print(movies_q3.columns)
print(movies_q3.shape)
```

```python
movies_q3
```

```python
# Q4 Determine the Netflix movies based on language, runtime and country
```

```python
movies_q4 = mov_ott.pivot(index='Title', columns='Netflix', values=['Language', 'Runtime', 'Country'])
```

```python
print(movies_q4.columns)
print(movies_q4.shape)
```

```python
movies_q4
```
